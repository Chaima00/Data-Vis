### Since the missing values represent almost 15% from all the dataset, we can just drop them

```python
df.dropna(inplace = True)
```

### Verify if there are missing values:

```python
sns.heatmap(df.isna(), yticklabels = False, cbar = True, cmap ='viridis');
```

### We removed all the rows (observations) that contain any missing value

```python
df.info()
```

### The prices are in string format

```python
df['1995_MED_PRICE_ALL_TYPES']
```

### Converting all the prices to float to perform more calculations like average, median...

```python
for i in df.columns:
    if 'PRICE' in i:
        df[i] = (df[i].str.replace(",", "")).apply(lambda x:float(x))
```

```python
df.info()
```

### Let's do some data visualisation

```python
fig,ax = plt.subplots(figsize=(8,6))
sns.distplot(df['1995_MED_PRICE_ALL_TYPES'],kde=True,ax=ax)
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Freq')
```

```python
sub_sample = df.sample(frac=0.01)
```

```python
sns.heatmap(sub_sample.corr(), cmap='Blues')
```

```python
fig,ax = plt.subplots(figsize=(8,6))
sns.boxplot(x = 'MSOA_NAME', y = '1995_MED_PRICE_ALL_TYPES',  data = df.sample(frac=0.01), ax=ax)
plt.title('City vs Price in 1995')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```
